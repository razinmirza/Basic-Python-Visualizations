**Problem 3.3A**

```python
laptop_df = pd.read_csv('D:\Documents\B412 Analytics\SEMESTER 2\BUS4023 Bus web\Week2\LaptopSalesJanuary2008(2).csv')
laptop_df
```

```python
laptop_df1 = laptop_df.rename(columns={'Store Postcode': 'Store_Postcode', 'Retail Price': 'Retail_Price'})
laptop_df1
```

```python
#Bar Chart showing the average Retail Price by each Store PostCode
ax = laptop_df1.groupby('Store_Postcode').mean().Retail_Price.plot(kind='bar')
ax.set_ylabel('Average Retail Price')
ax.set_xlabel('Store Postcode')
```

**Problem 3.3B**

```python
#Boxplots of Retail Prices by Store
ax1 = laptop_df1.boxplot(column='Retail_Price', by='Store_Postcode', rot=90)
ax1.set_ylabel('Retail Price')
ax1.set_xlabel('Store Postcode')
plt.suptitle('') # Suppress the titles
plt.title('')
```
